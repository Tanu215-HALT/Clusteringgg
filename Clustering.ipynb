1. What is unsupervised learning in the context of machine learning?

-  Unsupervised learning is a category of machine learning that aims to discover hidden patterns and inherent structures within data that has not been labeled or classified [1, 2].

Unlike supervised learning, there is no "correct" output or ground truth provided during the training phase. The algorithm works independently to infer the natural organization of the data itself [2].

The main goal of unsupervised learning is exploratory: to model the underlying probability distribution in the data to learn more about the dataset's composition.

Common Techniques and Tasks

Key tasks accomplished through unsupervised learning techniques include:

Clustering: This involves grouping similar data points together into clusters, where points within a cluster are more similar to each other than to points in other clusters [2]. Common algorithms include K-Means, hierarchical clustering, and DBSCAN.

Dimensionality Reduction: This is used to reduce the number of variables or features in a dataset while preserving meaningful information. This simplifies the data, reduces noise, and makes visualization easier. Principal Component Analysis (PCA) is a widely used example [2, 1].

Association Rule Learning: This method identifies strong relationships or associations between variables in large databases, such as determining which products are frequently purchased together in a supermarket [2].


2.How does K-Means clustering algorithm work

-  K-Means clustering is an unsupervised machine learning algorithm used to partition a dataset into K distinct, non-overlapping groups (clusters) based on similarity.

How K-Means Works:

Choose the number of clusters (K):
Decide how many clusters you want the algorithm to find.

Initialize centroids:
Randomly select K points from the dataset as the initial cluster centers (centroids).

Assign points to clusters:
For each data point, calculate the distance to each centroid. Assign the point to the cluster whose centroid is closest.

Update centroids:
For each cluster, recalculate the centroid by taking the mean of all points assigned to that cluster.

Repeat:
Repeat steps 3 and 4 until the cluster assignments no longer change or until a maximum number of iterations is reached.

Summary of the process:

The algorithm groups data points so that points in the same cluster are more similar to each other than to those in other clusters.
It minimizes the sum of squared distances between points and their respective cluster centroids.
Limitations:

You must choose K in advance.
K-Means assumes clusters are roughly spherical and of similar size.
Sensitive to initial centroid placement and outliers.
Example:
If you have data about customers (e.g., age, income), K-Means can help group customers with similar profiles together.

4.Explain the concept of a dendrogram in hierarchical clustering?

-  A dendrogram is a tree-like diagram that illustrates the arrangement of clusters produced by hierarchical clustering algorithms.

Key Points:

Hierarchical Clustering: This clustering method builds nested clusters by either merging (agglomerative) or splitting (divisive) them successively.
Dendrogram Visualization: The dendrogram visually represents how clusters are combined (or split) at each step. Each leaf (bottom) represents an individual data point, and branches show how clusters form as you move upward.
Height/Distance: The vertical axis (height) corresponds to the distance (or dissimilarity) at which clusters were joined. The shorter the vertical lines connecting clusters, the more similar those clusters are.
Cutting the Dendrogram: By "cutting" the dendrogram at a particular height, you can select a desired number of clusters.
Why Use a Dendrogram?

It helps visualize the process and structure of clustering.
It aids in deciding the optimal number of clusters by showing where large jumps in linkage distance occur.
Example

Suppose you have a dataset of animals with features (size, weight, habitat). Hierarchical clustering groups similar animals together, and the dendrogram shows these groupings and how closely related different clusters are.




4.What is the main difference between K-Means and Hierarchical Clustering?

-  The main difference between K-Means and Hierarchical Clustering lies in their approach to forming clusters and their output structure:

K-Means Clustering

Partitioning Method: Divides the data into K non-overlapping clusters by optimizing cluster centers (centroids).

Requires K: You must specify the number of clusters (K) before running the algorithm.

Flat Clustering: Produces a single partition of the data; no hierarchy or nested structure.

Efficiency: Generally faster and more scalable for large datasets.
Cluster Shape Assumption: Works best when clusters are spherical and of similar size.

Hierarchical Clustering

Hierarchical Method: Builds a tree of clusters (dendrogram) by either successively merging (agglomerative) or splitting (divisive) clusters.

No Need for K Upfront: You don’t need to specify the number of clusters before running; you can choose how many clusters you want by “cutting” the dendrogram at a desired level.

Nested Clustering: Produces a hierarchy of clusters, showing relationships at multiple levels.

Visualization: Dendrograms provide a visual way to understand cluster relationships.

Computational Cost: More expensive and less scalable for large datasets.



5.What are the advantages of DBSCAN over K-Means?

-  DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means clustering:

1. No Need to Specify Number of Clusters

DBSCAN: Automatically determines the number of clusters based on data density.

K-Means: Requires the user to specify K (the number of clusters) in advance.

2. Can Find Arbitrarily Shaped Clusters

DBSCAN: Can identify clusters of any shape (not just spherical), making it suitable for complex data distributions.

K-Means: Assumes clusters are spherical and equally sized, which may not fit real-world data.

3. Handles Noise and Outliers

DBSCAN: Effectively identifies and separates noise/outliers from clusters, labeling them as "noise."

K-Means: Assigns every point to a cluster, even if it is an outlier.

4. Works Well with Non-Linear Data

DBSCAN: Excels in datasets with non-linear boundaries between clusters.

K-Means: Struggles with data that has non-linear cluster boundaries.

5. No Assumption of Cluster Size

DBSCAN: Does not assume clusters are of similar size or density.

K-Means: Works best when clusters have similar sizes and densities.


6.When would you use Silhouette Score in clustering?

-  Silhouette Score is used in clustering to evaluate the quality of clusters and determine how well an object lies within its cluster compared to other clusters.

When to Use Silhouette Score:

To Choose the Optimal Number of Clusters (K):

When running clustering algorithms (like K-Means, Hierarchical, DBSCAN), you can calculate the silhouette score for different values of K (number of clusters).

The value of K that yields the highest average silhouette score typically indicates the best clustering solution.

To Assess Cluster Separation:

If you want to know how well-separated your clusters are, silhouette score provides an indication: values close to 1 mean well-separated clusters, values near 0 mean overlapping clusters, and negative values indicate misclassified points.

To Compare Different Clustering Algorithms:

If you run multiple clustering algorithms or use different parameters, silhouette score helps you compare which clustering result is better.


7.What are the limitations of Hierarchical Clustering?

-   1. Scalability (Computational Complexity)

Problem: It’s computationally expensive for large datasets.

Effect: Not suitable for datasets with thousands or millions of points.

 2. No Clear Objective Function

Unlike k-means (which minimizes variance), hierarchical clustering doesn’t optimize a specific objective.

This makes it harder to evaluate cluster quality mathematically.

 3. Irreversible Merging / Splitting

Once two clusters are merged (in agglomerative) or a cluster is split (in divisive), the decision cannot be undone, even if it later turns out to be suboptimal.

 4. Sensitive to Noise and Outliers

Outliers can cause distortion in the dendrogram and affect how clusters form.
A single outlier can create its own cluster or merge incorrectly.

 5. Choice of Distance Metric and Linkage Method

The outcome heavily depends on:

Distance metric (e.g., Euclidean, Manhattan, Cosine)

Linkage method (e.g., single, complete, average, Ward)

Different choices can lead to very different cluster structures.

 6. Difficulty Deciding the Number of Clusters

The dendrogram must be cut at a certain level to form clusters, but there’s no clear or automatic way to choose the optimal cut point.

 7. Assumes Hierarchical Structure

It forces data into a hierarchical shape, even if the true structure is non-hierarchical (e.g., overlapping or irregular clusters).

8.Why is feature scaling important in clustering algorithms like K-Means?

-   1. K-Means Uses Distance to Form Clusters
K-Means groups points based on how close they are to each other.
It minimizes the sum of squared distances between points and their cluster centers.

If one feature has much larger numerical values than another,
it will dominate the distance calculation — even if it’s not actually more important.

 2. Unscaled Features Cause Biased Clusters

For example:

Feature	Scale	Meaning

Height	150–190 cm	Small range

Income	30,000–200,000 ₹	Large range

If you don’t scale the features:

“Income” values dominate the distance computation

The algorithm will ignore the “Height” feature’s influence

Result → Poor or misleading clusters

 3. Scaled Features Ensure Equal Contribution

When you scale all features to the same range (e.g., 0–1 with Min-Max Scaling, or to mean 0 and variance 1 with Standardization),
each feature contributes equally to the distance metric.

This helps K-Means:

Detect clusters based on true patterns

Avoid bias toward features with large magnitudes

Improve cluster quality and interpretability



9.How does DBSCAN identify noise points?

-  DBSCAN (Density-Based Spatial Clustering of Applications with Noise) detects noise points, or outliers, by analyzing how densely data points are packed in a region.

 Key Idea:

It looks at each point’s neighborhood using two parameters:

ε (Epsilon): the distance that defines a point’s neighborhood.

MinPts: the minimum number of nearby points needed to consider a region dense.

 Classification of Points:

Core Point: Has at least MinPts neighbors within ε — starts or expands a cluster.

Border Point: Has fewer than MinPts neighbors but lies close to a core point — part of a cluster.

Noise Point: Has too few neighbors and isn’t close to any core point — treated as an outlier.

 In Simple Terms:

If a point is surrounded by many nearby points, it’s in a dense cluster.

If it’s isolated or lies in a sparse region, DBSCAN marks it as noise.

10. Define inertia in the context of K-Means?

-  Definition:
Inertia is the sum of squared distances between each data point and the centroid of the cluster it belongs to.

 Intuition:

Low inertia → points are close to their centroids → compact, well-formed clusters

High inertia → points are spread out → clusters may be less meaningful

 Use in Practice:

Inertia is used in the Elbow Method to help choose the optimal number of clusters (k) — where inertia stops decreasing significantly as k increases.




11.What is the elbow method in K-Means clustering?

-  The Elbow Method is a technique used in K-Means clustering to determine the optimal number of clusters (k) in a dataset.

 Idea:

It’s based on observing how inertia (the sum of squared distances of points from their cluster centroids) changes as you increase the number of clusters.

 Steps:

Run K-Means for a range of
k
k values (e.g., 1 to 10).
For each
k
k, calculate the inertia (within-cluster sum of squares).
Plot
k
k (x-axis) vs. inertia (y-axis).

Look for the “elbow” point — the point after which the decrease in inertia becomes smaller (the curve bends like an elbow).

 Interpretation:

Before the elbow: adding more clusters greatly reduces inertia (improves clustering).

After the elbow: adding clusters yields only minor improvements — likely overfitting.

The k value at the elbow point is chosen as the optimal number of clusters.

12. Describe the concept of "density" in DBSCAN?

-  In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), the concept of density is central to how clusters are formed.

Density in DBSCAN refers to how closely packed the data points are within a region.

A region is considered dense if there are enough points (at least MinPts) located within a specified distance (ε) from a point.

 Key Parameters:

ε (Epsilon): Defines the radius of a point’s neighborhood.

MinPts: The minimum number of points required within that ε-radius for the area to be considered dense.

 How Density Determines Clusters:

A Core Point lies in a dense region → it has ≥ MinPts neighbors within distance ε.

A Border Point lies close to a dense region → it has fewer than MinPts neighbors but is near a core point.

A Noise Point lies in a sparse region → it doesn’t meet either condition.

 Intuition:

High-density regions = clusters

Low-density regions = gaps or noise

DBSCAN connects points that are density-reachable — meaning they are part of a chain of neighboring dense regions — and treats these connected dense areas as clusters.

13. Can hierarchical clustering be used on categorical data?

-   Yes — hierarchical clustering can be used on categorical data, but not directly using standard distance measures like Euclidean distance.

It requires special distance metrics or data transformations to handle categorical variables properly.

Let’s break it down

 1. Why It’s Tricky

Hierarchical clustering normally depends on distance (or similarity) between points — e.g., Euclidean or Manhattan distance — which assume numeric and continuous features.

Categorical data (like “Red,” “Blue,” “Green” or “Yes,” “No”) doesn’t have a natural numerical order, so those metrics don’t make sense.

 2. How to Handle Categorical Data

There are two main approaches:

Option 1: Use specialized distance measures
You can compute dissimilarity using metrics that work for categorical data, such as:

Hamming distance: counts how many attributes differ between two items

Jaccard distance: compares the similarity between sets of attributes

Simple matching coefficient (SMC): proportion of matching categories

After computing this distance matrix, you can apply agglomerative hierarchical clustering (e.g., single, complete, average linkage).

Option 2: Encode categories numerically

Convert categories into numeric form using:

One-Hot Encoding (binary vectors)

Ordinal Encoding (if order matters)

 Note: After encoding, distance measures like Euclidean can be used — but be careful, as encoding can distort real relationships between categories.


3. Best Practices

For purely categorical data → use a distance metric for categorical variables.

For mixed data (numerical + categorical) → use Gower distance, which handles both types smoothly.

Use linkage methods like average or complete linkage for stable results.

14. What does a negative Silhouette Score indicate?

-  The Silhouette Score measures how well each point fits within its assigned cluster compared to other clusters.

A high positive score means the point is well grouped with its cluster.

A score near zero means it lies between two clusters.

A negative score means it’s closer to another cluster than to its own.

 Why Negative Scores Occur

The wrong number of clusters is chosen

The distance measure doesn’t suit the data

15. Explain the term "linkage criteria" in hierarchical clustering?

-  In hierarchical clustering, the linkage criteria (or linkage method) defines how the distance between two clusters is calculated when they are being merged.

When hierarchical clustering builds the cluster tree (dendrogram), it repeatedly merges the two closest clusters.

But — “closest” can mean different things depending on how you measure the distance between clusters, and that’s exactly what the linkage criterion decides.

| Linkage Type         | How Distance Is Calculated                                             | Description                                          |
| -------------------- | ---------------------------------------------------------------------- | ---------------------------------------------------- |
| **Single Linkage**   | Minimum distance between any two points (one from each cluster)        | Forms long, chain-like clusters; sensitive to noise. |
| **Complete Linkage** | Maximum distance between any two points (one from each cluster)        | Creates compact, well-separated clusters.            |
| **Average Linkage**  | Average distance between all pairs of points from the two clusters     | Balances between single and complete linkage.        |
| **Ward’s Linkage**   | Increase in total within-cluster variance when two clusters are merged | Often gives small, spherical, and balanced clusters. |

The choice of linkage affects:

The shape and size of clusters

The structure of the dendrogram

How sensitive the algorithm is to noise and outliers

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

K-Means can perform poorly on data with clusters of different sizes or densities because of the assumptions it makes about what clusters should look like.

1. K-Means Assumes Equal Cluster Size and Density
K-Means tries to minimize the distance between points and their cluster centroids.

It works best when clusters are:

Roughly spherical (round in shape)

Have similar sizes and equal density

When real clusters vary in size or density, K-Means struggles because:

The algorithm uses Euclidean distance, which treats all points equally, regardless of how dense or spread out the cluster is.

 2. Unequal Cluster Sizes

If one cluster is large and spread out, and another is small and dense:

The large cluster’s centroid can pull points away from the smaller one.

K-Means may split large clusters incorrectly or merge small clusters with larger ones.

 Example:

If one cluster has 100 points and another has 10, the large one dominates the centroid updates, leading to biased results.

 3. Unequal Densities

K-Means assigns each point to the nearest centroid, regardless of how dense the region is.

So:

Points in sparse regions may be grouped incorrectly with dense clusters.

Sparse clusters may be ignored or merged into nearby dense ones.

 4. Non-Spherical Shapes

K-Means assumes clusters are circular (spherical) in space.

If the true clusters are elongated, curved, or irregular, K-Means will misclassify points, since it can only divide space with straight lines (Voronoi boundaries).

17.  What are the core parameters in DBSCAN, and how do they influence clustering?

-  In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), the clustering process is mainly controlled by two core parameters — ε (Epsilon) and MinPts (Minimum Points).

These two parameters together decide how clusters form and how noise is detected.

1. ε (Epsilon) — Neighborhood Radius

Definition: The maximum distance within which DBSCAN considers two points as neighbors.

Influence:

A small ε → only points very close to each other are grouped → may result in many small clusters or more noise.

A large ε → more points fall within each neighborhood → may form fewer, larger clusters or even merge all points into one cluster.

 Think of ε as how far you “look around” a point to find neighbors.

2. MinPts — Minimum Number of Points

Definition: The minimum number of points required within the ε-radius for a region to be considered dense enough to form a cluster.

Influence:

A low MinPts → small or even random noise groups may be labeled as clusters.

A high MinPts → requires denser regions to form clusters → results in fewer but stronger clusters (and possibly more noise).

 Think of MinPts as how “crowded” a region must be to count as a cluster.

 | Case                  | Effect                                  |
| --------------------- | --------------------------------------- |
| Small ε + Low MinPts  | Many small clusters and scattered noise |
| Small ε + High MinPts | Sparse data → mostly noise              |
| Large ε + Low MinPts  | Few large, merged clusters              |
| Large ε + High MinPts | May ignore smaller clusters             |




18. How does K-Means++ improve upon standard K-Means initialization?

-  K-Means++ is an improved version of the K-Means initialization step that helps the algorithm find better starting centroids — leading to faster convergence and more accurate clusters.

1. Problem with Standard K-Means Initialization

In regular K-Means:

The initial centroids are chosen randomly.

Poor or unlucky choices can lead to:

Slow convergence

Suboptimal clusters

Different results each time you run it (unstable performance)

2. How K-Means++ Works

K-Means++ fixes this by choosing initial centroids more intelligently:

Pick the first centroid randomly from the data points.

For each remaining point, compute its distance from the nearest chosen centroid.

Choose the next centroid with probability proportional to the squared distance — points farther away from existing centroids are more likely to be chosen.

Repeat until k centroids are selected.

Then proceed with normal K-Means clustering.

3. Why It’s Better

Ensures initial centroids are well spread out.

Reduces the risk of poor clustering.

Leads to faster convergence and lower final inertia (tighter clusters).

Produces more consistent results across runs.


19. What is agglomerative clustering?

-  Agglomerative clustering is a type of hierarchical clustering that builds clusters in a bottom-up manner — starting with each data point as its own cluster and then merging them step by step based on similarity.

How It Works (Step-by-Step):

Start: Treat each data point as a single cluster.


→ If there are n points, you start with n clusters.

Compute distances: Calculate the distance (or similarity) between every pair of clusters using a linkage criterion (like single, complete, average, or Ward linkage).

Merge clusters: Combine the two closest clusters into one.

Update distances: Recalculate the distances between the new cluster and all other clusters.

Repeat: Continue merging until only one large cluster remains (or until a chosen number of clusters is reached).

Visualize: The process can be represented using a dendrogram, which shows how clusters were merged at each step.

20.What makes Silhouette Score a better metric than just inertia for model evaluation?

-  The Silhouette Score is often considered a better metric than inertia for evaluating clustering quality — especially when you want to compare how well-separated and meaningful your clusters actually are.

| Metric               | What it Measures                                                  | Limitation                                                                                            |
| -------------------- | ----------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------- |
| **Inertia**          | Compactness of clusters (how close points are to their centroids) | Only measures **within-cluster distance** — doesn’t consider how **far clusters are from each other** |
| **Silhouette Score** | Both **compactness** *and* **separation** between clusters        | Gives a more complete picture of cluster quality                                                      |


 2. Silhouette Score Considers Both Aspects

Silhouette Score looks at:

How close a point is to other points in its own cluster (cohesion)

How far it is from points in the nearest other cluster (separation)

This means it evaluates how distinct and well-separated clusters are — not just how tight they are.

3. Interpretable Range

Silhouette Score ranges from -1 to +1, which makes interpretation easy:

+1: Perfectly clustered — points are far from other clusters

0: Overlapping clusters — unclear boundaries

Negative: Misclassified points — likely in the wrong cluster

Inertia, on the other hand, has no fixed range, making it harder to compare between datasets.

4. Inertia Always Decreases as k Increases

Inertia will always go down when you add more clusters (even if it doesn’t improve real structure).

Silhouette Score avoids this trap — it often peaks at the optimal number of clusters, helping you choose the best k.

```python
21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot

# Import necessary libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
X, y_true = make_blobs(
    n_samples=400,        # number of data points
    centers=4,            # number of clusters
    cluster_std=0.6,      # standard deviation (controls spread)
    random_state=42       # for reproducibility
)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Visualize the clusters
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Plot cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.7, marker='X', label='Centroids')

plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot

 # Import required libraries
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)

# Step 2: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Step 3: Identify core and noise points
# In DBSCAN, label = -1 means the point is considered noise (outlier)
outliers = labels == -1
clusters = labels != -1

# Step 4: Visualize results
plt.figure(figsize=(8,6))
plt.scatter(X[clusters, 0], X[clusters, 1], c=labels[clusters], cmap='viridis', s=50, label='Clusters')
plt.scatter(X[outliers, 0], X[outliers, 1], c='red', s=80, marker='x', label='Outliers')

plt.title("DBSCAN Clustering on make_moons Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each
cluster

# Import necessary libraries
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Step 1: Load the Wine dataset
wine = load_wine()
X = wine.data

# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply K-Means clustering (let’s assume 3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Step 4: Print the size of each cluster
unique, counts = np.unique(labels, return_counts=True)
print("Cluster sizes:")
for cluster, size in zip(unique, counts):
    print(f"Cluster {cluster}: {size} samples")
```

```python
44.Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load the Iris dataset
data = load_iris()
X = data.data

# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Step 4: Compute silhouette scores
silhouette_avg = silhouette_score(X_scaled, labels)
sample_silhouette_values = silhouette_samples(X_scaled, labels)

print(f"Average Silhouette Score: {silhouette_avg:.3f}")

# Step 5: Plot silhouette coefficient for each sample
plt.figure(figsize=(8,6))
y_lower = 10
for i in range(3):
    # Aggregate silhouette scores for samples in cluster i
    cluster_silhouette_vals = sample_silhouette_values[labels == i]
    cluster_silhouette_vals.sort()

    cluster_size = cluster_silhouette_vals.shape[0]
    y_upper = y_lower + cluster_size

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, cluster_silhouette_vals,
                      alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * cluster_size, str(i))
    y_lower = y_upper + 10  # gap between clusters

plt.title("Silhouette Plot for Each Sample (Iris Dataset, K-Means Clustering)")
plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster Label")
plt.axvline(x=silhouette_avg, color="red", linestyle="--", label="Average Silhouette Score")
plt.legend()
plt.show()
```

```python
46.Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features)

# Import necessary libraries
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the Wine dataset
data = load_wine()
X = data.data
feature_names = data.feature_names

# Step 2: Convert to DataFrame (for Seaborn)
df = pd.DataFrame(X, columns=feature_names)

# Step 3: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Step 5: Add cluster labels to DataFrame
df['Cluster'] = labels

# Step 6: Visualize using Seaborn pairplot (first 4 features)
sns.pairplot(df.iloc[:, :5], hue='Cluster', palette='viridis', diag_kind='kde')
plt.suptitle("K-Means Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
 47.Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count

# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate noisy synthetic data
X, _ = make_blobs(n_samples=400, centers=3, cluster_std=1.0, random_state=42)
# Add random noise
rng = np.random.RandomState(42)
X = np.concatenate([X, rng.uniform(low=-10, high=10, size=(40, 2))])  # 40 noise points

# Step 2: Apply DBSCAN
db = DBSCAN(eps=0.8, min_samples=5)
labels = db.fit_predict(X)

# Step 3: Count clusters and noise points
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")

# Step 4: Visualize results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='plasma', s=40)
plt.title("DBSCAN Clustering with Noise")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

```

```python
48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters.

# Import necessary libraries
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Step 1: Load the Digits dataset
digits = load_digits()
X = digits.data

# Step 2: Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_embedded = tsne.fit_transform(X)

# Step 3: Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_embedded)

# Step 4: Visualize clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=labels, cmap='tab10', s=40)
plt.title("Agglomerative Clustering on Digits Data (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
